```python
# Checking if number_reviews (on listings) is according to what's on the reviews collection

pipeline = list(db.reviews.aggregate([
  {
    "$lookup": {
      "from": "listings",
      "localField": "_id",
      "foreignField": "_id",
      "as": "listing"
    }
  },
  
  {
    "$project": {
      "_id": 1,
      "number_of_reviews": "$listing.number_of_reviews",
      "num_reviews": { "$size": "$reviews" }
    }
  }
]))
count = 0
id_to_correct = []
value_to_correct = []

for doc in pipeline:
    doc['number_of_reviews'] = doc['number_of_reviews'][0]
    if doc['number_of_reviews'] != doc['num_reviews']:
        count += 1
        print(doc['_id'], doc['num_reviews'], doc['number_of_reviews'])
        id_to_correct.append(doc['_id'])
        value_to_correct.append(doc['num_reviews'])
       
print(count)
```

```python
# There are 244 listings that say they don't have any reviews - we'll update the number_of_reviews to 0 

for i, id in enumerate(id_to_correct):
    db.listingsAndReviews_new.update_many({'_id': id}, {'$set': {'number_of_reviews': value_to_correct[i]}})

for i, id in enumerate(id_to_correct):
    db.listings.update_many({'_id': id}, {'$set': {'number_of_reviews': value_to_correct[i]}})
```

```python
# To check if the last_review (in listingsAndReviews_new) is according to the reviews

pipeline = list(db.listingsAndReviews_new.aggregate([
  # unwind the reviews array
  { "$unwind": "$reviews" },
  # group by listing_id and find the max date
  {
    "$group": {
      "_id": "$reviews.listing_id",
      "maxDate": { "$max": "$reviews.date" }
    }
  },
  # lookup the corresponding listing and compare dates
  {
    "$lookup": {
      "from": "listingsAndReviews_new",
      "localField": "_id",
      "foreignField": "_id",
      "as": "listing"
    }
  },
  {
    "$project": {
      "_id": 1,
      "maxDate": 1,
      "last_review": { "$arrayElemAt": [ "$listing.last_review", 0 ] }
    }
  },
  # filter out documents where last_review is equal to maxDate
  {
    "$match": {
      "$expr": { "$ne": [ "$last_review", "$maxDate" ] }
    }
  },
  {
    "$project": {
      "_id": 1,
      "maxDate": 1,
      "last_review": 1,
      "is_different": { "$gt": [ "$maxDate", "$last_review" ] }
    }
  }
]))

pprint(pipeline)
pprint("Total number of documents: " + str(len(pipeline)))
```

### Question 11

Creating an 'calculate_average_price' function that calculates the average price for a given num_days, on a specific apartment

```python
def calculate_average_price(property_name, num_days):
    # Query the database for the given property
    property_doc = db.listings.find_one({"name": property_name})

    # Check if the property exists
    if property_doc is None:
        return None

    # Get the minimum and maximum number of nights
    min_nights = property_doc.get('minimum_nights')
    max_nights = property_doc.get('maximum_nights')

    # Check if num_days is within the allowed range
    if min_nights is not None and num_days < min_nights:
        return None
    if max_nights is not None and num_days > max_nights:
        return None

    # Get the price for the given number of days
    nightly_price = property_doc.get('price', {}).to_decimal()
    if nightly_price is None:
        return None

    # Convert the nightly_price to float
    nightly_price = float(nightly_price)

    # Calculate the average price per staying
    total_price = nightly_price * num_days
    average_price = total_price / num_days

    print("The average price for", property_doc.get('name'), " for ", num_days," has a daily average of ", average_price, " and a total price of ",total_price,".")
```
